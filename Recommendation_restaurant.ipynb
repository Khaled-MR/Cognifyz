```python
import numpy as np
import pandas as pd
```

```python
df  = pd.read_csv('Dataset .csv')
```

```python
df
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['Cuisines'].fillna(df['Cuisines'].mode()[0], inplace=True)
```

```python
df.info()
```

```python
df.columns.to_list()
```

```python
Cuisines = df['Cuisines'].value_counts()
Cuisines
```

```python
to_keep = Cuisines[Cuisines >= 50].index
df_cleaned = df[df['Cuisines'].isin(to_keep)]
```

```python
df_cleaned.info()
```

```python
def create_textual_representation(row):
    textual_representation =f"""
    Restaurant Name: {row['Restaurant Name']},
    City: {row['City']},
    Cuisines: {row['Cuisines']},
    Average Cost for two: {row['Average Cost for two']},
    Currency: {row['Currency']},
    Has Table booking: {row['Has Table booking']},
    Has Online delivery: {row['Has Online delivery']},
    Is delivering now: {row['Is delivering now']},
    Price range: {row['Price range']},
    Aggregate rating: {row['Aggregate rating']},
    Rating text: {row['Rating text'] }
    """
    return textual_representation
```

```python
df_cleaned['textual_representation'] = df_cleaned.apply(create_textual_representation, axis=1)
```

```python
print(df_cleaned['textual_representation'].values[0])
```

```python
import faiss
import requests
import numpy as np

dim = 4096
index = faiss.IndexFlatL2(dim)

X = np.zeros((len(df_cleaned['textual_representation']), dim),dtype='float32')
```

```python
for i , representation in enumerate(df_cleaned['textual_representation']):
    if i % 2:
       print('processed',str(i),'instances')
    res = requests.post('http://localhost:11434/api/embeddings',
                      json={
                          'model':'llama2',
                          'prompt':representation
                      }
                      )
    embedding = res.json()['embedding']
    X[i] = np.array(embedding)


index.add(np.array([embedding]))
```

```python
faiss.write_index(index,'index')
```

```python
index = faiss.read_index('index')
```

```python
favorite_movie = df_cleaned.iloc[800]
```

```python
res = requests.post('http://localhost:11434/api/embeddings',
                      json={
                          'model':'llama2',
                          'prompt':favorite_movie['textual_representation']
                      }
                      )
```

```python
embedding = np.array([res.json()['embedding']], dtype='float32')
D,I = index.search(embedding,5)
```

```python
best_matchs = np.array(df_cleaned['textual_representation'])[I.flatten()]
```

```python
for match in best_matchs:
    print("NEXT MOVIE")
    print(match)
    print('-'*100)
```
